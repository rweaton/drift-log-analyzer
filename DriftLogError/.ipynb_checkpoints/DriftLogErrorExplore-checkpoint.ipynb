```python
# major module imports
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d
from importlib import reload
```

```python
class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)

# define a well path shape: here, a spiral
def spiral_gen(radius, pitch, z):
    
    x = radius*np.cos(pitch*z)
    y = radius*np.sin(pitch*z)
    
    return (x, y, z)

# function to transform the well path from cylindrical
# (r, theta, z) to downhole coordinates (path-length,
# incline, azimuth)
def cyl2dh_xform(x, y, z):
    
    delta_x = np.diff(x)
    delta_y = np.diff(y)
    delta_z = np.diff(z)
    delta_l = np.sqrt(delta_x**2 + delta_y**2 + delta_z**2)
    
    az = np.arctan2(delta_y, delta_x)
    inc = np.arccos(np.divide(delta_z, delta_l))
    l =  np.cumsum(delta_l)
    
    return (l, inc, az)

R = 4.
a = 2*np.pi*.01
z = np.linspace(0., -200., num=801, endpoint=True)

rx, ry, rz = spiral_gen(R, a, z)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(rx-R, ry, rz)

l, inc, az = cyl2dh_xform(rx, ry, rz)
```

```python
fig2, ax2 = plt.subplots(nrows=1, ncols=2)
ax2[0].plot(l, (180./np.pi)*az)
ax2[1].plot(l, inc)
```

```python
# Add Gaussian noise to inc and az arrays; added noise
# intended to simulate measurement error in probe
# sensors.
# Specify variance of noise distrib. for each coordinate
sigma_l = 0.05*(l[1] - l[0])
sigma_az = 5*1.8*np.pi/180.
sigma_inc = 5*1.8*np.pi/180.

(sigma_l, sigma_az, sigma_inc)
```

```python
# Produce arrays of random floats for the two coordinates
seed_l = np.random.rand(*l.shape)
seed_az = np.random.rand(*az.shape)
seed_inc = np.random.rand(*inc.shape)

# Add noise to the two signals using the inverseCDF function
noisy_l = nps.inverseCDF_norm(seed_l, l, sigma_l)
noisy_az = nps.inverseCDF_norm(seed_az, az, sigma_az)
noisy_inc = nps.inverseCDF_norm(seed_inc, inc, sigma_inc)

# Inspect noisy signals by plotting.
fig3, ax3 = plt.subplots(nrows=1, ncols=3)
ax3[0].plot(l, noisy_l)
ax3[1].plot(l, (180./np.pi)*noisy_az)
ax3[2].plot(l, (180./np.pi)*noisy_inc)
```

```python
DelNorth, DelEast, DelTVD = dpcl.BalancedTangentialMethod(l, inc, az, AngleUnits='radians')
fig4 = plt.figure()
ax4 = fig4.add_subplot(111, projection='3d')
ax4.plot(np.cumsum(DelNorth), np.cumsum(DelEast), np.cumsum(DelTVD), color='gray', linestyle='--')

DelNorth, DelEast, DelTVD = dpcl.AverageAngleMethod(noisy_l, noisy_inc, noisy_az, AngleUnits='radians')
ax4.plot(np.cumsum(DelNorth), np.cumsum(DelEast), np.cumsum(DelTVD), color='black')

DelNorth, DelEast, DelTVD = dpcl.BalancedTangentialMethod(noisy_l, noisy_inc, noisy_az, AngleUnits='radians')
ax4.plot(np.cumsum(DelNorth), np.cumsum(DelEast), np.cumsum(DelTVD), color='cyan')

DelNorth, DelEast, DelTVD = dpcl.RadiusOfCurvatureMethod(noisy_l, noisy_inc, noisy_az, AngleUnits='radians')
ax4.plot(np.cumsum(DelNorth), np.cumsum(DelEast), np.cumsum(DelTVD), color='red')

DelNorth, DelEast, DelTVD = dpcl.MinimumCurvatureMethod(noisy_l, noisy_inc, noisy_az, AngleUnits='radians')
ax4.plot(np.cumsum(DelNorth), np.cumsum(DelEast), np.cumsum(DelTVD), color='green')
```

```python
reload(dpcl)
def errorBootstrap(processFunc, nReps, signals, sigmas, **params):
    # Size check: each signal requires an associated sigma 
    # (standard deviation of noise to be added)
    if signals.shape[0] == sigmas.shape[0]:
        output = np.empty((nReps, *signals.shape))
        sigIndices = range(0, signals.shape[0])
        for i in range(0, nReps):
            seeds = np.random.rand(*signals.shape)
            noisySigs = np.array([nps.inverseCDF_norm(seeds[j,:], signals[j,:], sigmas[j]) 
                                  for j in sigIndices])
            output[i,:,:] = np.array([processFunc(*noisySigs, **params)])
    else:
        output = -1
        print('Error: lengths of signal and sigma arrays do not match!')
    return output

nReps = 500
bsDels = errorBootstrap(dpcl.MinimumCurvatureMethod, nReps, np.array([l, inc, az]), 
                         np.array([sigma_l, sigma_inc, sigma_az]))
```

```python
paths = np.array([np.array([np.cumsum(bsDels[i,0,:]), np.cumsum(bsDels[i,1,:]), 
             np.cumsum(bsDels[i,2,:])]) for i in range(0, nReps)])
```

```python
fig5 = plt.figure()
ax5 = fig5.add_subplot(111, projection='3d')
ax5.plot(rx-R, ry, rz, linewidth=3, color='red')
for i in range(0,nReps):
    #path = (np.cumsum(bsDels[i,0,:]), np.cumsum(bsDels[i,1,:]), 
    #         np.cumsum(bsDels[i,2,:]))
    ax5.plot(*paths[i,:,:], color='gray', alpha=.2)
```

```python
j = 750
covMat = np.cov(paths[:,:,j].transpose())
lambdas, pcs = np.linalg.eig(covMat)
aVecs = np.dot(np.diag(np.sqrt(lambdas)), np.transpose(pcs))

meanPath = np.array([np.mean(paths[:, 0, j]), np.mean(paths[:, 1, j]), 
                     np.mean(paths[:, 2, j])])

arrows = [Arrow3D([meanPath[0], v[0]], [meanPath[1], v[1]], 
                  [meanPath[2], v[2]], mutation_scale=20, 
                  lw=2, arrowstyle="-|>", color="r") for v in aVecs]
arrowCoords = [[[meanPath[0], meanPath[0]+v[0]], 
               [meanPath[1], meanPath[1]+v[1]],
               [meanPath[2], meanPath[2]+v[2]]] for v in aVecs]

fig6 = plt.figure()
ax6 = fig6.add_subplot(111, projection='3d')
ax6.scatter(paths[:, 0, j], paths[:, 1, j], paths[:, 2, j], alpha=0.2)
ax6.scatter(*meanPath, marker='+', c='k', s=200, alpha=1.)
ax6.scatter(rx[j]-R, ry[j], rz[j], marker='+', color='r', s=200, alpha=1.)
for arrow in arrowCoords:
    ax6.plot(*arrow, c='black', linewidth=1)
#for arrow in arrows:
#    ax.add_artist(arrow)
```
